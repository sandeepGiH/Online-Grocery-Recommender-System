### LIGHT FM HYBRID MODEL FOR E-GROCERY REC. SYSTEM

#### Advantages of Light FM 

A hybrid matrix factorization model by Lyst.

LightFM is a hybrid matrix factorisation model representing users and items as linear combinations of their content features’ latent factors. The model outperforms both collaborative and content-based models in cold-start or sparse interaction data scenarios (using both user and item metadata), and performs at least as well as a pure collaborative matrix factorisation model where interaction data is abundant.

LightFM model learns embeddings (latent representations in a high-dimensional space) for users and items in a way that encodes user preferences over items. When multiplied together, these representations produce scores for every item for a given user; items scored highly are more likely to be interesting to the user

#### Benefit of LightFM python library

The biggest benefit of lightfm library is that it implements WARP (Weighted Approximate-Rank Pairwise) loss for implicit feedback learning-to-rank.

WARP works like these:

For a given (user, positive item pair), sample a negative item at random from all the remaining items. Compute predictions for both items; if the negative item’s prediction exceeds that of the positive item plus a margin, perform a gradient update to rank the positive item higher and the negative item lower. If there is no rank violation, continue sampling negative items until a violation is found.

If you found a violating negative example at the first try, make a large gradient update: this indicates that a lot of negative items are ranked higher than positives items given the current state of the model, and the model must be updated by a large amount. If it took a lot of sampling to find a violating example, perform a small update: the model is likely close to the optimum and should be updated at a low rate.

```python
from lightfm.cross_validation import random_train_test_split
from lightfm.evaluation import auc_score, precision_at_k, recall_at_k
from lightfm import LightFM
from scipy.sparse import csr_matrix
```

```python
cust_id = list(custitem.index)
cust_dict = {}
counter = 0
for i in cust_id:
    cust_dict[i] = counter
    counter += 1

item_list = list(custitem.columns)
item_dict ={}
counter = 0
for i in item_list:
        item_dict[counter] =i
        counter += 1
    
# convert to csr matrix

# Light FM needs data to be in matrix..
cust_item_csr = csr_matrix(custitem.values)
cust_item_csr
```

#### no_components (int, optional) 
– the dimensionality of the feature latent embeddings.

#### learning_rate (float, optional)
– initial learning rate for the adagrad learning schedule.

#### loss (string, optional) 
– one of (‘logistic’, ‘bpr’, ‘warp’, ‘warp-kos’): the loss function.


```python
# Model Training
model = LightFM(loss = 'warp',random_state=2016, learning_rate=0.9, no_components = 150, user_alpha=0.000005)
model = model.fit(cust_item_csr,epochs = 100, num_threads = 16, verbose = False)
```

## Item recommendation to a user

##### we want to show a user, items he might be interested in buying/viewing based on his/her interactions done in the past. Typical industry examples for this are like “Deals recommended for you”..

```python
def sample_Recommendation_user(cust_id,threshold = 0, nrec_items = 10):
    n_cust,n_items = custitem.shape
    cust_x = cust_dict[cust_id]
    
    scores = pd.Series(model.predict(cust_x,np.arange(n_items)))
    scores.index = custitem.columns
    
    scores = list(pd.Series(scores.sort_values(ascending = False).index))
    
    known_items = list(pd.Series(custitem.loc[cust_id,:][custitem.loc[cust_id,:] > threshold].index).sort_values(ascending=False))
    scores = [x for x in scores if x not in known_items]
    return_score_list = scores[0:nrec_items]
    known_items = list(pd.Series(known_items))
    scores = list(pd.Series(return_score_list)) 
            
    print('\n recommended items:')
    return scores
```

## User recommendation to a item
#### we can recommend a list of users specific to a particular item. Example of such cases is when you are running a promotion on an item and want to run an e-mail campaign around this promotional item to only 10,000 users who might be interested in this item.

```python
def sample_recommendation_item(item_id,number_of_user):
    n_users,n_items = custitem.shape
    x = np.array(list(item_dict.keys()))
    scores = pd.Series(model.predict(np.arange(n_users),np.repeat(x.searchsorted(item_id),n_users)))
    users_list = list(custitem.index[scores.sort_values(ascending = False).head(number_of_user).index])
    return users_list
```

## Item recommendation to items

#### we will discuss how we can recommend a list of items specific to a particular item. This kind of models will help you to find similar/related items or items which can be bundled together. 
#### Typical industry use case for such models are in cross-selling and up-selling opportunities on product page like “Products related to this item”, “Frequently bought together”, “Customers who bought this also bought this” and “Customers who viewed this item also viewed”.

```python
from sklearn.metrics.pairwise import cosine_similarity

def item_item_Recommendation(item,n_items):
    #create_item_embed_dist_matrix
    item_norm_sparse = csr_matrix(model.item_embeddings)
    similarities = cosine_similarity(item_norm_sparse)
    item_embed_dist_matrix = pd.DataFrame(similarities)
    item_embed_dist_matrix.columns = custitem.columns
    item_embed_dist_matrix.index = custitem.columns
    # recommend
    recommend_items = list(pd.Series(item_embed_dist_matrix.loc[item,:].sort_values(ascending = False).head(n_items+1).index[1:n_items+1]))
    print('Item similar to above item:')
    return recommend_items
```
